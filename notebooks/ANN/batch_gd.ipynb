```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4], [7,9,5], [6,10,6], [5,12,7]], columns=['CGPA', 'IQ', 'LPA'])
df
```

```python
x = df.iloc[:,:-1]
y = df['LPA']
```

```python
def batch_gd(x, y):
    n_samples, n_features = x.shape
    weights = np.ones(x.shape[1])*0.01
    bias = 0
    lr = 0.1
    
    for i in range(1000):
        y_hat = output(np.dot(x, np.transpose(weights)) + bias) 
        error = (y-y_hat)
        dw = -(1/n_samples) * np.dot( np.transpose(x), error )
        weights = weights - lr*dw
        
    return weights
    
    
def output(z):
        return np.where(z > 0, 1, 0)
    
    
                   
```

```python
batch_gd(x, y)
```
