# **MULTI-LAYER-PERCEPTRON (MLP)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
df = pd.DataFrame(([1, 20, 23,0], [2, 30, 10,1], [4, 40, 3, 0]), columns=['num','ID', 'IQ', 'JOB'])
x = df

x
```

```python
def mlp_first_layer(x):
    nodes = int(input("Enter number of nodes in the first layer of MLP: "))
    weights = np.ones((nodes, x.shape[1]))*0.1
    bias = np.ones(nodes)*0.001
    epochs = 100
    all_neuron = {}
    for i in range(nodes):
        all_neuron['neuron ' + str(i+1)] = np.dot(x, np.transpose(weights[i])) + bias[i]
            
    return all_neuron

    
```

```python
values =mlp_first_layer(x)
```

```python
ans = list(values.keys())
neuron = values[ans[0]]
neuron = neuron.reshape(1,3)
neuron
```

```python
def mlp_last_layer(all_neuron, nodes):
    last_node = int(input("Enter the number of last node: "))
    bias = np.ones(last_node)*0.01
    weights_fin = np.ones(nodes, last_node)*0.1
    for i in range(nodes):
        neuron = all_neuron[ans[i]]
        neuron = neuron.reshape(1,3)
        output = np.dot(weights_fin, neuron)
    return output
```
