### **Reading the document and preprocessing**

```python
with open("/mnt/c/Users/Subham Pathak/Desktop/AI/DEEP_LEARNING/notebooks/RNN/datasets/1661-0.txt", 'r') as file:
    docs = file.read()
    file.close()
    print(docs)
```

```python
words = docs.split("\n")

sentence = [s.strip() for s in words if s.strip()]
sentence
```

### **Splitting the data into training and testing datas**

```python
from sklearn.model_selection import train_test_split
x_Train, x_test = train_test_split(sentence, random_state=41, test_size = 0.2)
```

```python
print(f"The x_train and x_test is a python list not an numpy array of len x_train = {len(x_Train)}, x_test = {len(x_test)}.")
```

### **Converting the text into integer values and padding them to equal length**

```python
token = Tokenizer(oov_token='hehe')
token.fit_on_texts(x_Train)
```

```python
x_train_token = token.texts_to_sequences(x_Train)
x_test_token = token.texts_to_sequences(x_test)
print(x_test_token)
print(x_train_token)
```

```python
from keras.utils import pad_sequences
```

```python
x_train_pad = pad_sequences(x_train_token, padding='post')
x_test_pad = pad_sequences(x_test_token, padding='post')

x_train_pad
```
