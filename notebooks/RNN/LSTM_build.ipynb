```python
class The_forget_gate:
    
    
    def __init__(self, x):
        self.x = x
        self.print()
        
    def sigmoid(self, Z):
        return 1/(1+np.exp(-Z))
    
    def main(self):
        
        ht = np.random.random(3)
        conc = np.concatenate([self.x, ht])
        length = len(conc)
        weights = np.random.rand(3, length)
        
        return self.sigmoid(np.dot(weights, conc))
    
    def print(self):
        print("Class created successfully.")
    
```
