```python
import numpy as np
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.datasets import imdb
from keras.layers import Dense, Input, Flatten, SimpleRNN, Embedding
from keras import Sequential, callbacks
from keras.utils import pad_sequences
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data()
```

```python
print(f"Shape of x_test is : {x_test.shape}")
print(f"Shape of x_train is : {x_train.shape}")
```

```python
len(x_train[3])
```

```python
x_train = pad_sequences(x_train, padding="post", maxlen=50)
x_test = pad_sequences(x_test, padding="post", maxlen=50)
```

```python
model = Sequential()

model.add(Input(shape=(50,1)))
model.add(SimpleRNN(30), return_sequences = False )
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x_train, y_train, batch_size=32, epochs=10, validation_data=(x_test, y_test), callbacks=callbacks)
```
